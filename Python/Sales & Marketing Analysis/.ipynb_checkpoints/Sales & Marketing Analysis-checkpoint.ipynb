## Creating database & tables 

```python
# Connect to SQLite database (or create it)
conn = sqlite3.connect('DEPI__InsideOut_Project.db')
cursor = conn.cursor()

# 1. Customers table
cursor.execute('''
CREATE TABLE IF NOT EXISTS customers (
    Customer_ID TEXT PRIMARY KEY,
    Gender TEXT,
    Location TEXT
);
''')

# 2. Products table
cursor.execute('''
CREATE TABLE IF NOT EXISTS products (
    Product_ID TEXT PRIMARY KEY,
    Product_Type TEXT
);
''')

# 3. Orders table
cursor.execute('''
CREATE TABLE IF NOT EXISTS orders (
    Date_ID TEXT,
    Order_ID TEXT PRIMARY KEY,
    Customer_ID TEXT,
    Product_ID TEXT,
    Order_Status TEXT,
    No_of_products INTEGER,
    Product_cost INTEGER,
    Product_price INTEGER,
    Total_Price INTEGER,
    Dollar_exchange_rate FLOAT,
    FOREIGN KEY(Product_ID) REFERENCES Products(Product_ID),
    FOREIGN KEY(Customer_ID) REFERENCES Customers(Customer_ID),
    FOREIGN KEY(Date_ID) REFERENCES date(Date_ID)
);
''')

# 4. Date table
cursor.execute('''
CREATE TABLE IF NOT EXISTS date (
    Date_ID TEXT PRIMARY KEY,
    Date TEXT
);
''')

# 5. Marketing table
cursor.execute('''
CREATE TABLE IF NOT EXISTS marketing (
    Date_ID TEXT,
    Content_type TEXT,
    Post_type TEXT,
    likes INTEGER,
    comments INTEGER,
    shares INTEGER,
    views INTEGER,
    Platform_ID INTEGER,
    FOREIGN KEY(Date_ID) REFERENCES date(Date_ID),
    FOREIGN KEY(Platform_ID) REFERENCES Platforms(Platform_ID)
);
''')

# 6. Platforms table
cursor.execute('''
CREATE TABLE IF NOT EXISTS platforms (
    Platform_ID TEXT PRIMARY KEY,
    Platform_name TEXT
);
''')

#7. cities table
cursor.execute('''
CREATE TABLE IF NOT EXISTS cities (
    Platform_ID TEXT,
    City TEXT,
    Percentage FLOAT,
    FOREIGN KEY(Platform_ID) REFERENCES Platforms(Platform_ID)
);
''')

#8. countries table
cursor.execute('''
CREATE TABLE IF NOT EXISTS countries (
    Platform_ID TEXT,
    Country TEXT,
    Percentage FLOAT,
    FOREIGN KEY(Platform_ID) REFERENCES Platforms(Platform_ID)
);
''')

#8. gender table
cursor.execute('''
CREATE TABLE IF NOT EXISTS gender (
    Platform_ID TEXT,
    Gender TEXT,
    Percentage FLOAT,
    FOREIGN KEY(Platform_ID) REFERENCES Platforms(Platform_ID)
);
''')

#9. age table
cursor.execute('''
CREATE TABLE IF NOT EXISTS age (
    Platform_ID TEXT,
    Age TEXT,
    Percentage FLOAT,
    FOREIGN KEY(Platform_ID) REFERENCES Platforms(Platform_ID)
);
''')

conn.commit()
#conn.close()
```

### Outliers

```python
# Calculate the IQR and detect outliers for views in each platform

facebook_data = df_marketing[df_marketing['Platform_ID'] == 111]
    
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'views'
Q1 = facebook_data['views'].quantile(0.25)
Q3 = facebook_data['views'].quantile(0.75)
    
# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1
    
# Determine outliers: values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers = facebook_data[(facebook_data['views'] < (Q1 - 1.5 * IQR)) | (facebook_data['views'] > (Q3 + 1.5 * IQR))]
    
# Display results
print(outliers)
```

```python
# Calculate the IQR and detect outliers for views in each platform

instagram_data = df_marketing[df_marketing['Platform_ID'] == 222]
    
# Calculate Q1 (25th percentile) and Q3 (75th percentile) for 'views'
Q1 = instagram_data['views'].quantile(0.25)
Q3 = instagram_data['views'].quantile(0.75)
    
# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1
    
# Determine outliers: values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers_2 = instagram_data[(instagram_data['views'] < (Q1 - 1.5 * IQR)) | (instagram_data['views'] > (Q3 + 1.5 * IQR))]
    
# Display results
print(outliers_2)
```

## Marketing analysis

### 1. Relationship between content type and metrics (likes, comments, shares, views) on Facebook, Instagram, and TikTok

```python
# Connect to the database
conn = sqlite3.connect('DEPI_Project.db')

#SQLLit Query to join tables together 

query = '''
SELECT 
    platforms.Platform_name, 
    marketing.Content_type,
    marketing.likes,
    marketing.comments,
    marketing.shares,
    marketing.views
FROM 
    marketing
JOIN 
    platforms ON marketing.Platform_ID = platforms.Platform_ID
'''
Content_type_sql = pd.read_sql_query(query, conn)
print(Content_type_sql)
```

```python
#Aggregating median values for metrics based on content type & platform 
Content_type_median = Content_type_sql.groupby(['Content_type', 'Platform_name'])[['likes', 'comments', 'shares', 'views']].median().sort_values(by='views', ascending=False)
Content_type_median
```

```python
#Pivoting median values for metrics based on content type & platform 
Content_type_pivot = Content_type_sql.pivot_table(index='Content_type', columns='Platform_name', values=['likes', 'comments', 'shares', 'views'], aggfunc='median')
Content_type_pivot
```

```python
import matplotlib.pyplot as plt

# Define the color palette for different platforms
colors = ['#c4d8f3', '#ffcc9e', 'lavender', '#fffc9e', '#c4e8f3']

# Create a figure with two rows and two columns of square plots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# 1. Plot for Comments
Content_type_pivot[['comments']].plot(kind='bar', ax=axs[0, 0], color=colors, width=0.75, legend=False)
axs[0, 0].set_title('Comments by Content type and Platform', fontsize=14)
axs[0, 0].set_xlabel('Content Type', fontsize=10)
axs[0, 0].set_ylabel('Median Comments', fontsize=10)
axs[0, 0].tick_params(axis='x', rotation=45)

# 2. Plot for Likes
Content_type_pivot[['likes']].plot(kind='bar', ax=axs[0, 1], color=colors, width=0.75, legend=False)
axs[0, 1].set_title('Likes by Content Type and Platform', fontsize=14)
axs[0, 1].set_xlabel('Content Type', fontsize=10)
axs[0, 1].set_ylabel('Median Likes', fontsize=10)
axs[0, 1].tick_params(axis='x', rotation=45)

# 3. Plot for Shares
Content_type_pivot[['shares']].plot(kind='bar', ax=axs[1, 0], color=colors, width=0.75, legend=False)
axs[1, 0].set_title('Shares by Content Type and Platform', fontsize=14)
axs[1, 0].set_xlabel('Content Type', fontsize=10)
axs[1, 0].set_ylabel('Median Shares', fontsize=10)
axs[1, 0].tick_params(axis='x', rotation=45)

# 4. Plot for Views
Content_type_pivot[['views']].plot(kind='bar', ax=axs[1, 1], color=colors, width=0.75, legend=False)
axs[1, 1].set_title('Views by Content Type and Platform', fontsize=14)
axs[1, 1].set_xlabel('Content Type', fontsize=10)
axs[1, 1].set_ylabel('Median Views', fontsize=10)
axs[1, 1].tick_params(axis='x', rotation=45)

# Add a common legend
fig.legend(['Comments', 'Likes', 'Shares', 'Views'], loc='upper right', fontsize=12)

# Adjust layout
plt.tight_layout(rect=[0, 0, 0.85, 0.95])  # Adjust rect to add space for the common legend

# Show the plots
plt.show()
```

### 5. Total number of Orders by Month

```python
#SQL query to link date with sales to count no of orders in each month

query = '''
SELECT 
    strftime('%m', d.Date) AS Month,  -- Extract month from the date
    COUNT(o.Order_ID) AS Total_Orders -- Count distinct orders for each month
FROM 
    orders o
LEFT JOIN 
    date d ON o.Date_ID = d.Date_ID
GROUP BY Month
ORDER BY Month
'''

# Execute the query and fetch the results
cursor.execute(query)
result = cursor.fetchall()

# Load the results into a pandas DataFrame for better visualization
df_month = pd.DataFrame(result, columns=['Month', 'Total_Orders'])

# Display the DataFrame
print(df_month)
```

```python
Orders_month = df_month.pivot_table(index='Month', values='Total_Orders')
Orders_month
```

### 6. Total Sales and Total Units by Year

```python
# sql query to link tables together 

query = '''
SELECT 
    strftime('%Y', d.Date) AS Year,  -- Extract the year from the date
    SUM(o.Total_Price) AS Total_Sales,  -- Sum of total sales
    SUM(o.No_of_products) AS Total_Units_Sold  -- Sum of total units sold
FROM 
    orders o
LEFT JOIN 
    date d ON o.Date_ID = d.Date_ID
GROUP BY Year
ORDER BY Year
LIMIT 3  -- Limit to only the last three years
'''

# Execute the query and fetch the results
cursor.execute(query)
result = cursor.fetchall()

# Load the results into a pandas DataFrame
year_sales = pd.DataFrame(result, columns=['Year', 'Total_Sales', 'Total_Units_Sold'])

# Display the DataFrame to verify the data
print(year_sales)
```

```python
Pivot_year = year_sales.pivot_table(index='Year', values=['Total_Sales','Total_Units_Sold'])
Pivot_year
```

```python
# SQL query with UNION of LEFT JOIN and RIGHT JOIN equivalent
query = '''
-- First part: LEFT JOIN to get all orders, even if there's no matching marketing data
SELECT 
   strftime('%Y-%m', d.Date) AS Month,   
   SUM(DISTINCT(COALESCE(m.views, 0))) AS Total_Views,  
   COUNT(DISTINCT(o.Order_ID)) AS Total_Orders    
FROM 
    orders o
LEFT JOIN 
    marketing m ON o.Date_ID = m.Date_ID  
LEFT JOIN 
    date d ON o.Date_ID = d.Date_ID 
GROUP BY MONTH
    '''

# Execute the query and fetch the results
cursor.execute(query)
result = cursor.fetchall()

# Load the results into a pandas DataFrame for better visualization
df_marketing_sales = pd.DataFrame(result, columns=['Month', 'Total_Views', 'Total_Orders'])

# Display the DataFrame
print(df_marketing_sales)
```

```python
# Create a pivot table with Month as the index and values as Total_Views and Total_Orders
marketing_sales_piv = df_marketing_sales.pivot_table(index='Month', values=['Total_Views', 'Total_Orders'], aggfunc='sum')

# Display the pivot table
marketing_sales_piv 
```

```python
#visualize total views with total sales

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Create a bar plot for Total_Views
ax1.bar(marketing_sales_piv.index, marketing_sales_piv['Total_Orders'], color='lightblue', label='Total Orders', alpha=0.7)

# Create a second y-axis for Total_Orders
ax2 = ax1.twinx()
ax2.plot(marketing_sales_piv.index, marketing_sales_piv['Total_Views'], color='orange', marker='o', label='Total Views')

# Adding titles and labels
ax1.set_title('Total Views and Total Orders by Month', fontsize=16)
ax1.set_xlabel('Month', fontsize=14)
ax1.set_ylabel('Total Orders', fontsize=14)
ax2.set_ylabel('Total Views', fontsize=14)

# Adding grid and legends
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

### 2. The relationship between count of posts each month & total no of orders

```python
#Inserting post id column in marketing table to count no of posts by it

import sqlite3
import uuid

# Connect to SQLite database
conn = sqlite3.connect('DEPIII_Project.db')
cursor = conn.cursor()

# SQL Query to check if the Post_ID column exists (it should after ALTER)
cursor.execute('''PRAGMA table_info(marketing)''')
columns = cursor.fetchall()
if 'Post_ID' not in [col[1] for col in columns]:
    cursor.execute('''ALTER TABLE marketing ADD COLUMN Post_ID TEXT AUTO_INCREMENT;''')

# Function to update Post_IDs with unique values
cursor.execute('SELECT rowid FROM marketing')  # Get row IDs to ensure uniqueness
rows = cursor.fetchall()

for row in rows:
    post_id = str(uuid.uuid4())  # Generate unique Post_ID
    cursor.execute('''
        UPDATE marketing 
        SET Post_ID = ? 
        WHERE rowid = ?''', (post_id, row[0]))

print("Post_IDs have been successfully added and updated.")
```

```python
#SQL query to join post id with order id to calculate total orders

query = '''
SELECT 
    d.Date AS Month,  -- Extract year-month for grouping
    COUNT(DISTINCT m.Post_ID) AS Post_Count,  -- Count of distinct posts by Post_ID
    COUNT(DISTINCT o.Order_ID) AS Total_Orders  -- Count of distinct orders by Order_ID
FROM 
    orders o
LEFT JOIN 
    marketing m ON o.Date_ID = m.Date_ID  
LEFT JOIN 
    date d ON o.Date_ID = d.Date_ID 
GROUP BY MONTH
'''

# Execute the query and fetch the results
cursor.execute(query)
result = cursor.fetchall()

# Load the results into a pandas DataFrame
count_orders = pd.DataFrame(result, columns=['Month', 'Post_Count', 'Total_Orders'])

# Display the DataFrame to verify the data
print(count_orders)
```

```python
count_orders_pivot = count_orders.pivot_table(index='Month', values=['Post_Count', 'Total_Orders'], aggfunc='sum')

# Display the pivot table
count_orders
```

```python
#visualization

# Set the figure size
plt.figure(figsize=(14, 7))

# Create a bar plot for Post_Count
ax1 = count_orders_pivot['Post_Count'].plot(kind='bar', color='lightblue', alpha=0.7, position=0, width=0.4, label='Post Count')

# Create a second y-axis for the line plot (Total_Orders)
ax2 = ax1.twinx()
ax2.plot(count_orders_pivot.index, count_orders_pivot['Total_Orders'], color='orange', marker='o', label='Total Orders')

# Adding titles and labels
ax1.set_title('Post Count and Total Orders by Month', fontsize=16)
ax1.set_xlabel('Month', fontsize=14)
ax1.set_ylabel('Post Count', fontsize=14)
ax2.set_ylabel('Total Orders', fontsize=14)

# Adding grid and legends
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
